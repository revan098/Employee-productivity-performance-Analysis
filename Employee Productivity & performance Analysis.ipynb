```python
#Employee Count by Department
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Department', order=df['Department'].value_counts().index, palette='viridis')
plt.title('Number of Employees per Department')
plt.xticks(rotation=45)
plt.show()
```

```python
#Gender Distribution
plt.figure(figsize=(5,5))
df['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightblue','lightpink','gray'])
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```

```python
#Average performance by department
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Department', y='Performance_Score', estimator='mean', ci=None, palette='crest')
plt.title('Average Performance Score by Department')
plt.xticks(rotation=45)
plt.show()
```

```python
#Department wise salary distrubtion
plt.figure(figsize=(10,6))
sns.boxplot(data=df,x='Department',y='Monthly_Salary',palette='coolwarm')
plt.title('Monthely Salary Distribution by department')
plt.xticks(rotation=45)
plt.show()
```

```python
#Relationship B/W Working Hours & Performance
plt.figure(figsize=(10,6))
sns.scatterplot(data=df,x='Work_Hours_Per_Week',y='Performance_Score',hue='Department',alpha=1)
plt.title('Working Hours vs Performance Score')
plt.show()
```

```python
#Comverting Hiredate and showing Trends
df['Hire_Year']=df['Hire_Date'].dt.year
plt.figure(figsize=(10,6))
df['Hire_Year'].value_counts().sort_index().plot(kind='bar',color='teal')
plt.title('Employed Hired per Year')
plt.xlabel('year')
plt.ylabel('Number of Employee')
plt.show()
```

```python
#Training Hours and performance
plt.figure(figsize=(10,6))
sns.scatterplot(data=df,x='Training_Hours',y='Performance_Score',hue='Department',alpha=1)
plt.title('Training Hours vs Performance')
plt.show()
```

```python
#Resignation Count by Department
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Department', hue='Resigned', palette='Set2')
plt.title('Resignation Count by Department')
plt.xticks(rotation=45)
plt.show()
```

```python
#employee satisfaction and performance
plt.figure(figsize=(10,6))
sns.lineplot(data=df, x='Employee_Satisfaction_Score', y='Performance_Score', hue='Department', alpha=0.6)
plt.title('Employee Satisfaction vs Performance')
plt.show()
```

```python
#Correlation Heatmap (All Numeric Columns)
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), cmap='coolwarm', annot=True, fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Traning Hours And Promotion
plt.figure(figsize=(8,6))
sns.regplot(data=df, x='Training_Hours', y='Promotions', scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title("Training Hours vs Promotions", fontsize=14)
plt.xlabel("Training Hours")
plt.ylabel("Number of Promotions")
plt.show()
```
